```python
import pandas as pd
import gzip
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#caminho do Arquivo
caminho_arquivo = "/title.basics.tsv.gz"

#Lê arquivo compactado usando gzip
with gzip.open(caminho_arquivo, 'rt') as arquivo:
  #leitura do arquvivo TSV
  data_frame = pd.read_csv(arquivo, delimiter='\t')

#exibi data-frame
print(data_frame)
```

```python
#visualiza as 100 primeiras linhas do data-frame.
print(data_frame.head(100))
```

```python
#Verifica informações básicas
print(data_frame.info())
```

```python
#Explora estatístias descritivas
print(data_frame.describe())
```

```python
#filta dados com condicionais
filtered_data = data_frame[data_frame['genres'] == 'Comedy']
```

```python
#conta valores únicos em um coluna
unique_values = data_frame['genres'].nunique()
print(unique_values)
```

```python
#verificar a distribuição dos tipos de títulos
type_counts = data_frame['titleType'].value_counts()
print(type_counts)
```

```python
#Gráfico de barras
type_counts.plot(kind='bar')
plt.xlabel('Tipo de Título')
plt.ylabel('Frequência')
plt.title('Distribuição dos Tipos de Títulos')
plt.show()
```

```python
#Analisa os gêneros mais populares
genre_counts = data_frame['genres'].value_counts()
print(genre_counts.head(10))

#Gráfico de Barras
top_genres = genre_counts.head(10)
top_genres.plot(kind='bar')
plt.xlabel('Gênero')
plt.ylabel('Frequência')
plt.title('Top 10 gêneros mais populares')
plt.show()
```
